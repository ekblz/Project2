```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/nicepeopleproject/ml_linear_regression/master/data/data.csv'
df = pd.read_csv(url)
df
```

```python
df.columns.str.replace(' ', '_').str.lower() # делаем названия консистентными
df.columns = df.columns.str.replace(' ', '_').str.lower()
df.columns = df.columns.str.lower().str.replace(' ', '_')
df
```

```python
n = len(df) # разбиваем данные на три части
n_train = int(len(df) * 0.6)
n_val = int(len(df) * 0.2)
n_test = n - n_val - n_train
```

```python
df_val = df.iloc[:n_val] # данные сгруппированы
df_test = df.iloc[n_val: n_val+n_test]
df_train = df.iloc[n_val+n_test:]
```

```python
idx = np.arange(n) # хотим данные перемешать
np.random.seed(2) # чтобы результаты были воспроизводимы
np.random.shuffle(idx)
```

```python
df_train = df.iloc[idx[:n_train]] # данные перемешаны
df_val = df.iloc[idx[n_train:n_train+n_val]]
df_test = df.iloc[idx[n_train+n_val:]]
```

```python
df_train = df_train.reset_index(drop=True) # индексы меняются
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True) 
df_train
```

```python
y_train = np.log1p(df_train.msrp.values)
y_val = np.log1p(df_val.msrp.values)
y_test = np.log1p(df_test.msrp.values)
```

```python
del df_train['msrp'] # чтобы случайно не использовать цену для предсказания цены
del df_val['msrp']
del df_test['msrp']
```

```python
df_train = df_train[list(df_train.dtypes[df_train.dtypes != object].index)].fillna(0)
df_test = df_test[list(df_test.dtypes[df_test.dtypes != object].index)].fillna(0)
# замена пустых ячеек на 0
```

```python
df_train.isnull().sum() 
df_test.isnull().sum() 
```

```python
df_train = df_train.values
df_test = df_test.values
```

```python
w = np.dot(df_train.T, df_train) # формула
w = np.linalg.inv(w)
w = np.dot(w, df_train.T)
w1 = np.dot(w, y_train)
```

```python
pred = np.dot(df_test, w1)
```
